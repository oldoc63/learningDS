### Sublime Limes' Line Graphs

In this project, you will be acting as a data analyst for an online lime retailer called Sublime Limes. People all over the world can use this product to get the freshest, best-quality limes delivered to their door. One of the product managers at Sublime Limes would like to gain insight into the customers and their sales habits. Using Matplotlib, you'll create some different line graphs to communicate this information effectively.

#### Understand the data and set up subplots

1. We have provided some data in different lists. Look through these list and try to understand what each one represents.

```python
from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]

persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]

blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

2. Create a figure of width 12 and height 8.

```python
plt.figure(figsize=(12, 8))
```

3. We are going to make two charts in one figure, laid out side-by-side. In other words, the figure will have one row and two columns. Write the command to create the left subplot. Save this subplot in a variable called ax1.

```python
plt.figure(figsize=(12, 8))

ax1 = plt.subplot(1,2,1)
```

4. Write the command to create the right subplot.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)

ax2 = plt.subplot(1,2,2)
```

#### Page visits over time

5. In the left subplot, we are going to plot the total page visits over the past year as a line.

    First, let's create the list of x-values, which is `range(len(months))`. Store this in a variable called `x_values`.

    Make sure this happens after the line where you created ax1, but before the line where you've created ax2.

6. Plot the total page visits against these x_values as a line.

7. Give the line markers that will help show each month as a distinct value.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)

x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')

ax2 = plt.subplot(1,2,2)
```

8. Label the x-axis and y-axis with descriptive titles of what they measure.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')

plt.xlabel('Month')
plt.ylabel('Visit Per Month')

ax2 = plt.subplot(1,2,2)
```

9. Set the x-axis ticks to be the x_values.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visit Per Month')

ax1.set_xticks(x_values)

ax2 = plt.subplot(1,2,2)
```

10. Label the x-axis tick labels to be the names stored in the months list

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visit Per Month')
ax1.set_xticks(x_values)

ax1.set_xticklabels(months)

ax2 = plt.subplot(1,2,2)
```

#### Plotting multiple lime species

11. In the subplot on the right, we are going to plot three lines on the same set of axes. The x-values for all three lines will correspond to the months, so we can use the list of x_values we used fot the last plot.

On one plot, create the three lines:
- Number of key limes sold vs x_values
- Number of Persian limes sold vs x_values
- Number of blood limes sold vs x_values

Make sure this happens after the line where you created ax2, so that it goes in the subplot on the right.

12. Add a legend to differentiate the lines, labeling each lime species.

13. Set the x-axis to be the x_values, and the tick labels to be the months list.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visit Per Month')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1,2,2)

plt.plot(x_values, key_limes_per_month)
plt.plot(x_values, persian_limes_per_month)
plt.plot(x_values, blood_limes_per_month)

plt.legend(['key', 'persian', 'blood'])

ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
```

#### Labeling and saving

14. Add a title to each of the two plots you've created, and adjust the margins to make the text you've added look better.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visit Per Month')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

plt.title('Page visits over last year')

ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month)
plt.plot(x_values, persian_limes_per_month)
plt.plot(x_values, blood_limes_per_month)
plt.legend(['key', 'persian', 'blood'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)

plt.title('Species of Limes sold by month')
```

15. Now, save your figure as png with a descriptive file name.

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Visit Per Month')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title('Page visits over last year')
ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month)
plt.plot(x_values, persian_limes_per_month)
plt.plot(x_values, blood_limes_per_month)
plt.legend(['key', 'persian', 'blood'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.title('Species of Limes sold by month')

plt.savefig('Sublime Limes Graph.png')
```
