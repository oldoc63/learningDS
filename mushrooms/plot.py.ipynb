```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load the data
df = pd.read_csv('mushroom_data.csv')
print(df.head())
```

```python
print(df.shape)
```

```python
# list of all column headers
columns = df.columns.tolist()
print(columns)
```

```python
for column in columns:
    #print(column)
    sns.countplot(df[column])
    plt.show()
    plt.clf()
```
