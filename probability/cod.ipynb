## Sampling Cod Fish Population

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv(r'C:\Users\varic\Documents\learningDS\probability\cod_population.csv')
population = population['Cod_Weight']
sample_size = 50
sample_means = []
```

```python
for i in range(500):
    samp = np.random.choice(population, sample_size, replace=False)
    # calculate mean here
    this_sample_mean = np.mean(samp)
    # append here
    sample_means.append(this_sample_mean)
```

```python
sns.histplot(sample_means, stat='density')
plt.title("Sampling Distribution of the mean")
plt.xlabel('Weight (lbs)')
plt.show()
```
