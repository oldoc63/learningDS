# Load, Clean, & Store Data

```python
import pandas as pd

# read in the csv file
df = pd.read_csv('Exercise-Data-Unclean.csv')

display(df)
```

```python
# drop any rows with missing data
df = df.dropna()
display(df)
```

```python
# remove any duplicates
df = df.drop_duplicates()
display(df)
```

```python
# replace any invalid data
df = df.replace(450, 45)
df = df.replace('Rowing', 'Rower')
display(df)

# write the cleaned data to a new Excel file
df.to_excel('Exercise-Data-Clean.xlsx', index=False)
```
