# Present Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_csv('Presentation-Data.csv')

# Filter the data to select data for one month
jan_df = df[df['Month'] == 'January']

# Create a bar chart
jan_df.plot(kind='bar', x='Product', y='Sales')
plt.title("January Sales")
plt.show()
```

```python
# Filter the data to select only one product
jeans_df = df[df['Product'] == 'Jeans']

# Create a line chart
jeans_df.plot(kind='line', x='Month', y='Sales')
plt.title("Jeans Sales")
plt.show()
```
